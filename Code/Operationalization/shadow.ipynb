```python
# function to round observations values
def round_observations(x):
    return float(round(x, 0))

# importing the prophet prediction dataset
prophet_prediction_dataset = pd.read_csv('../../Docs/Assets/prophet_prediction.csv')

# converting the index to DateTime
prophet_prediction_dataset['ds'] = pd.to_datetime(prophet_prediction_dataset['ds'])

# getting only the index and ApparentTemperature column
# apparent_temperature_prediction_dataset = prophet_prediction_dataset[['ds','yhat_upper']].copy()
apparent_temperature_prediction_dataset = prophet_prediction_dataset[['ds','yhat']].copy()

# renaming the columns
# apparent_temperature_prediction_dataset.rename(columns = {'ds':'timestamp', 'yhat_upper':'ApparentTemperature'}, inplace = True)
apparent_temperature_prediction_dataset.rename(columns = {'ds':'timestamp', 'yhat':'ApparentTemperature'}, inplace = True)

# setting the index
apparent_temperature_prediction_dataset.set_index('timestamp', inplace=True)

# rounding ApparentTemperature column values
apparent_temperature_prediction_dataset['ApparentTemperature'] = apparent_temperature_prediction_dataset['ApparentTemperature'].apply(round_observations)
```

```python
'''
/*
 * Copyright 2010-2017 Amazon.com, Inc. or its affiliates. All Rights Reserved.
 *
 * Licensed under the Apache License, Version 2.0 (the "License").
 * You may not use this file except in compliance with the License.
 * A copy of the License is located at
 *
 *  http://aws.amazon.com/apache2.0
 *
 * or in the "license" file accompanying this file. This file is distributed
 * on an "AS IS" BASIS, WITHOUT WARRANTIES OR CONDITIONS OF ANY KIND, either
 * express or implied. See the License for the specific language governing
 * permissions and limitations under the License.
 */
 '''

# importing the necessary libraries
from AWSIoTPythonSDK.MQTTLib import AWSIoTMQTTShadowClient
import logging
import time
import json


# setting default/initial device shadow state
powered_on = "False"

# setting the AWS IoT and Thing certificates and parameters
SHADOW_PROPERTY = 'powered-on'
ENDPOINT = "[OMITTED FOR SECURITY PURPOSES]"
CLIEND_ID = "[OMITTED FOR SECURITY PURPOSES]"
PATH_TO_CERTIFICATE = "[OMITTED FOR SECURITY PURPOSES]"
PATH_TO_PRIVATE_KEY = "[OMITTED FOR SECURITY PURPOSES]"
PATH_TO_AMAZON_ROOT_CA_1 = "[OMITTED FOR SECURITY PURPOSES]"

# setting the AWS IoT shadow client and communication parameters
host = ENDPOINT
rootCAPath = PATH_TO_AMAZON_ROOT_CA_1
certificatePath = PATH_TO_CERTIFICATE
privateKeyPath = PATH_TO_PRIVATE_KEY
port = 8883
useWebsocket = False
thingName = CLIEND_ID
clientId = CLIEND_ID


# Custom Shadow callback
def customShadowCallback_Update(payload, responseStatus, token):
    # payload is a JSON string ready to be parsed using json.loads(...)
    # in both Py2.x and Py3.x
    if responseStatus == "timeout":
        # print("Update request " + token + " time out!")
        pass
    if responseStatus == "accepted":
        payloadDict = json.loads(payload)
        # print("~~~~~~~~~~~~~~~~~~~~~~~")
        # print("Update request with token: " + token + " accepted!")
        # print("~~~~~~~~~~~~~~~~~~~~~~~\n\n")
    if responseStatus == "rejected":
        pass
        # print("Update request " + token + " rejected!")

# function to GET the device shadow state on AWS
def customShadowCallback_Get(payload, responseStatus, token):

    if responseStatus == "timeout":
        print("GET request " + token + " time out!")
    if responseStatus == "accepted":
        payloadDict = json.loads(payload)

    print("~~~~~~~~~~~~~~~~~~~~~~~")
    print("GET request with token: " + token + " accepted!")
    print("~~~~~~~~~~~~~~~~~~~~~~~\n\n")

    if responseStatus == "rejected":
        print("GET request " + token + " rejected!")


# Configure logging
logger = logging.getLogger("AWSIoTPythonSDK.core")
logger.setLevel(logging.INFO)  # DEBUG
streamHandler = logging.StreamHandler()
formatter = logging.Formatter(
    '%(asctime)s - %(name)s - %(levelname)s - %(message)s')
streamHandler.setFormatter(formatter)
logger.addHandler(streamHandler)


# configuring the AWS IoT shadow client
myAWSIoTMQTTShadowClient = AWSIoTMQTTShadowClient(clientId)
myAWSIoTMQTTShadowClient.configureEndpoint(host, port)
myAWSIoTMQTTShadowClient.configureCredentials(
    rootCAPath, privateKeyPath, certificatePath)

# AWSIoTMQTTShadowClient configuration
myAWSIoTMQTTShadowClient.configureAutoReconnectBackoffTime(1, 32, 20)
myAWSIoTMQTTShadowClient.configureConnectDisconnectTimeout(10)  # 10 sec
myAWSIoTMQTTShadowClient.configureMQTTOperationTimeout(5)  # 5 sec

# added this new line to handle off-line messages
myAWSIoTMQTTShadowClient.getMQTTConnection().configureOfflinePublishQueueing(20)

# Connect to AWS IoT
myAWSIoTMQTTShadowClient.connect()

# Create a deviceShadow with persistent subscription
deviceShadowHandler = myAWSIoTMQTTShadowClient.createShadowHandlerWithName(
    thingName, True)


# iteratting over dataset
for i in range(len(apparent_temperature_prediction_dataset['ApparentTemperature'])):

    # getting the current row value of the dataset
    predicted_temperature = apparent_temperature_prediction_dataset['ApparentTemperature'][i]

    # setting the desired state of the device shadow based on the predicted temperature value on the row
    powered_on = 'True' if predicted_temperature >= 24 else 'False'

    # printing ACIoT logo in digital ASCII art format
    print_aciot_logo()
    print('')

    # printing the current state of the device shadow with a digital ASCII art format
    print_ac_on(predicted_temperature) if powered_on == 'True' else print_ac_off(predicted_temperature)

    # mouting the device shadow state in JSON format
    json_desired = {
        "state": {
            "desired": {
                "welcome": "aws-iot",
                "powered-on": powered_on
            },
            "reported": {
                "welcome": "aws-iot",
                "powered-on": "False"
            }
        }
    }

    # updating the device shadow state
    deviceShadowHandler.shadowUpdate(json.dumps(json_desired), customShadowCallback_Update, 10)

    # sleeping for a second to avoid connection conflicts
    time.sleep(1)
```
